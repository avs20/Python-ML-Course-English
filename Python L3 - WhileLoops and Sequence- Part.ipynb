2. continue the loop in between. 

### exit the loop in between

````python
count = 1 
while count > 0 :
    num = 2 ** count 
    count = count + 1
    if num >= 1000:
        break
        
print(num )
````

### 2 continue the loop in betwee 

IT does not exit the loop but send the execution to the start of the loop 

````python
count = 1 
while count <= 10 :
    num = 2 ** count 
    count = count + 1
    if num >= 1000:
        continue
        
print(num )
````
